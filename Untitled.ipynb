```python
import agb.agbrom
```

```python
with open('proj.pmp.constants') as f:
    specieses = eval(f.read())['species']['values']
```

```python
with open('pokedex_order.csv') as f:
    new_order = f.read().splitlines()
```

```python
violet = agb.agbrom.Agbrom('bld/violet.gba')
```

```python
pokedex_order_offset = 0x14AE174
species_to_order = {}
for i, species in enumerate(new_order):
    species_to_order[species] = 0, i + 1

for i, species in enumerate(specieses[1:]):
    if not species in species_to_order:
        species_to_order[species] = 0, 0
    _, new = species_to_order[species]
    old = violet.u16(pokedex_order_offset + 2 * i)
    species_to_order[species] = old, new
    
```

```python
# Output the new C-Array
for species in specieses[1:]:
    print(f'\t{species_to_order[species][1]}, // {species}')
```

```python
# Collect specieses with changed id
id_changed = set()
for species in specieses[1:]:
    old, new = species_to_order[species]
    if old != new:
        id_changed.add(species)
```

```python
for species in species_to_order:
    old, new = species_to_order[species]
    if old >= 416: print(old, species)
```
