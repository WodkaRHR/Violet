```python
import os
```

```python
files = [f for f in os.listdir('.') if f.endswith('.flags')]
```

```python
for fn in files:
    new_flags = []
    with open(fn, 'r') as f:
        content = f.read()
        print('asdf', fn, content)
        flags = content.splitlines()[0].split(' ')
        for flag in flags:
            flag, value = flag[:2], flag[2:]
            if flag == '-V': 
                new_flags.append('-m ' + value)
            elif flag == '-G':
                new_flags.append('-g voicegroup' + value)
            elif flag == '-P':
                new_flags.append('-p ' + value)
            else:
                raise RuntimeError(f'Unknown flag {flag}')
    with open(fn, 'w+') as f:
        f.write(' '.join(new_flags))
        
    print(flags, new_flags)
        
```
